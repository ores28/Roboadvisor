```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.stattools import pacf

# Load dataset
df = pd.read_csv("gibf1_stable_sections.csv")

# Store results
section_results = []

# Loop through all sections (1 to 26)
for section in range(1, 44):
    df_section = df[df['Stable_Section'] == section]

    if 'closePrice' not in df_section.columns:
        print(f"Skipping section {section} (No 'closePrice' column)")
        continue

    df_section = df_section[['closePrice']].dropna()

    if len(df_section) < 10:
        print(f"Skipping section {section} (Not enough data)")
        continue

    print(f"\n🔹 Processing Section {section}...")

    # Calculate PACF to determine the best lag
    pacf_values = pacf(df_section['closePrice'], nlags=10, method='ols')
    best_k = np.argmax(np.abs(pacf_values[1:])) + 1  # Ignore the first value (lag 0)

    # Normalize data
    scaler = MinMaxScaler()
    close_prices = df_section['closePrice'].values.reshape(-1, 1)
    close_prices_scaled = scaler.fit_transform(close_prices)

    # Create input-output pairs using lag k
    def create_sequences(data, k):
        X, y = [], []
        for i in range(len(data) - k):
            X.append(data[i:i + k].flatten())  # Ensure 2D array shape
            y.append(data[i + k])
        return np.array(X), np.array(y)

    X, y = create_sequences(close_prices_scaled, best_k)

    # Ensure data is 2D
    if X.ndim == 1:
        X = X.reshape(-1, best_k)
    if y.ndim == 1:
        y = y.reshape(-1, 1)

    # Split into training and testing sets (80% train, 20% test)
    split_index = int(len(X) * 0.8)
    X_train, X_test = X[:split_index], X[split_index:]
    y_train, y_test = y[:split_index], y[split_index:]

    # Ensure X_train and X_test have correct shapes
    if X_train.shape[0] == 0 or X_test.shape[0] == 0:
        print(f"⚠️ Not enough data for training/testing in Section {section}. Skipping...")
        continue

    # Step 2: Define ANN Model
    model = Sequential([
        Dense(64, activation='relu', input_shape=(best_k,)),
        Dense(32, activation='relu'),
        Dense(1)  # Output layer
    ])

    model.compile(optimizer='adam', loss='mse')

    # Step 3: Train Model
    model.fit(X_train, y_train, epochs=200, batch_size=16, verbose=0, validation_data=(X_test, y_test))

    # Step 4: Evaluate Model
    y_pred = model.predict(X_test, verbose=0)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)

    print(f"📉 RMSE for Section {section}: {rmse:.5f}")

    # Step 5: Forecast Next 10 Steps
    forecast_scaled = []
    
    # Get last valid sequence
    last_sequence = X[-1].reshape(1, best_k)  # Ensure it's a 2D array

    for _ in range(10):
        next_pred = model.predict(last_sequence, verbose=0)[0][0]  # Predict next step
        forecast_scaled.append(next_pred)

        # Shift window for next prediction
        last_sequence = np.roll(last_sequence, -1)
        last_sequence[0, -1] = next_pred  # Append predicted value

    # Convert forecasted values back to original scale
    forecast = scaler.inverse_transform(np.array(forecast_scaled).reshape(-1, 1)).flatten()

    # Store results
    section_results.append({"Section": section, "Best_k": best_k, "RMSE": rmse})

    # Step 6: Plot actual vs forecast
    plt.figure(figsize=(8, 4))
    plt.plot(df_section.index, df_section['closePrice'], label="Actual", color="blue")
    plt.plot(range(len(df_section), len(df_section) + 10), forecast, label="Forecast (ANN)", color="red", linestyle="dashed")
    plt.title(f"NSSAR (ANN) - Section {section} | RMSE: {rmse:.5f}")
    plt.xlabel("Time")
    plt.ylabel("Close Price")
    plt.legend()
    plt.show()

# Step 7: Save results to CSV
results_df = pd.DataFrame(section_results)
results_df.to_csv("gibf1_nssar_ann_results.csv", index=False)

print("\n✅ NSSAR (ANN) Processing Complete! Results saved to 'gibf_nssar_ann_results.csv'.")
```

```python
section_predictions = pd.read_csv("gibf1_nssar_ann_results.csv")

# Compute Scaled RMSE (Normalization)
rmse_values = section_predictions["RMSE"].values
rmse_min = np.min(rmse_values)
rmse_max = np.max(rmse_values)
scaled_rmse = (rmse_values - rmse_min) / (rmse_max - rmse_min)

# Compute iFMII (Mean of Scaled RMSE)
ifmii = np.mean(scaled_rmse)

# Add the Scaled RMSE column
section_predictions["Scaled_RMSE"] = scaled_rmse  

# Rank iFMII values in descending order
section_predictions["iFMII_rank"] = section_predictions["Scaled_RMSE"].rank(ascending=False, method="min")

# Compute p-value
total_predictions = len(section_predictions)
section_predictions["p_value_nonlinear"] = section_predictions["iFMII_rank"] / total_predictions

# Append the overall iFMII as a new row
ifmii_row = pd.DataFrame({
    "Section": ["Overall iFMII"],
    "Best_k": [None],  
    "RMSE": [None],  
    "Scaled_RMSE": [None],  
    "iFMII_rank": [None],  
    "p_value_nonlinear": [None],  
    "iFMII": [ifmii]  
})

# Concatenate the new row to the existing DataFrame
section_predictions = pd.concat([section_predictions, ifmii_row], ignore_index=True)

# Save back to the same CSV file
section_predictions.to_csv("gibf1_nssar_ann_results.csv", index=False)

print(f"\n✅ iFMII Value: {ifmii:.5f} (Appended to 'gibf_nssar_ann_results.csv')")
```
